```python
import time
import random
import matplotlib.pyplot as plt
def bubblesort(arr):
  n = len(arr)
  for i in range(n-1):
    for j in range(0,n-1-i):
      if arr[j]>arr[j+1]:
        arr[j],arr[j+1] = arr[j+1],arr[j]
```

```python
#Function to calculate size list and time list and finally plot
def calc_and_plot(arr,case):
  #calculating t for different n and creating time list (t)
  n = list(range(0,10001,1000))
  t = list()
  for x in n:
    temp = arr[0:x]
    start = time.time()
    bubblesort(temp)
    end = time.time()
    t.append(end-start)
  #plotting graph
  plt.plot(n, t, label = case)
  plt.xlabel('Size of array')
  plt.ylabel('Time in seconds')
  plt.title('Size vs Time')
  plt.legend()
  #plt.show()

#create rand array of size 10000
arr_rand = list()
for x in range(10000):
  arr_rand.append(random.randrange(10000))
arr_sorted = sorted(arr_rand) # list(range(1000)) try for testing
arr_worstcase = sorted(arr_rand,reverse=1) # sorted(list(range(1000)),reverse=1) try for testing

calc_and_plot(arr_rand, "Average")
calc_and_plot(arr_sorted, "Best")
calc_and_plot(arr_worstcase, "Worst")
plt.show()
```
